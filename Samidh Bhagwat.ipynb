# Outliers check 

```python
# Function to detect outliers using IQR and return columns with outliers
def columns_with_outliers(loan_data):
    outlier_columns = []
    for column in loan_data.select_dtypes(include=['number']).columns:
        Q1 = loan_data[column].quantile(0.25)
        Q3 = loan_data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        if any((loan_data[column] < lower_bound) | (loan_data[column] > upper_bound)):
            outlier_columns.append(column)
    return outlier_columns

# Get columns with outliers
outlier_columns = columns_with_outliers(loan_data)

# Print columns with outliers
print("Columns with outliers:")
print(outlier_columns)
```

```python
# Function to calculate upper bound for columns with outliers using IQR
def calculate_upper_bound(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    return upper_bound

# Get upper bound values for columns with outliers
outlier_columns_upper_bounds = {}
for column in loan_data.select_dtypes(include=['number']).columns:
    Q1 = loan_data[column].quantile(0.25)
    Q3 = loan_data[column].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = calculate_upper_bound(loan_data, column)
    outliers = loan_data[loan_data[column] > upper_bound][column]
    if not outliers.empty:
        outlier_columns_upper_bounds[column] = upper_bound

# Print columns with outliers and their upper bound values
for column, upper_bound in outlier_columns_upper_bounds.items():
    print(f"Column '{column}': Upper Bound = {upper_bound}")
```

```python
#Outliers might affect the overall analysis and hence we need to treat them , 
#replacing them with median is a good option for now
# Function to replace outliers with median using IQR
def replace_outliers_with_median(loan_data):
    Q1 = loan_data.quantile(0.25)
    Q3 = loan_data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Replace outliers with the median of the column
    loan_data = loan_data.apply(lambda x: loan_data.median() if x < lower_bound or x > upper_bound else x)
    return loan_data
```

```python
# Choose the column for which you want to replace outliers with median
column_name = 'loan_amnt'  
column_name1 = 'funded_amnt'
column_name2 = 'funded_amnt_inv'
column_name3 = 'int_rate'
column_name4 = 'installment'
column_name5 = 'annual_inc'

# Replace outliers with median 
loan_data[column_name] = replace_outliers_with_median(loan_data[column_name])
loan_data[column_name1] = replace_outliers_with_median(loan_data[column_name1])
loan_data[column_name2] = replace_outliers_with_median(loan_data[column_name2])
loan_data[column_name3] = replace_outliers_with_median(loan_data[column_name3])
loan_data[column_name4] = replace_outliers_with_median(loan_data[column_name4])
loan_data[column_name5] = replace_outliers_with_median(loan_data[column_name5])
```

```python
#Creating a function to Plot relationship between multiple independent varaibles and the dependant variable
def plot_bar_chart(x_var, y_var):
    """
    Plot a bar chart with specified x and y variables.

    Parameters:
        x_var (str): Name of the variable to be plotted on the x-axis.
        y_var (str): Name of the variable to be plotted on the y-axis.

    Returns:
        None
    """
    plt.figure(figsize=(8, 6))
    sns.barplot(x=x_var, y=y_var, data=loan_data)
    plt.title(f'Bar Chart: {y_var} across {x_var}', fontsize=14)
    plt.xlabel(x_var.capitalize(), fontsize=12)
    plt.ylabel(y_var.capitalize(), fontsize=12)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Calculate the total number of loans and default loans per bin
bin_counts = loan_data['loan_amount_bin'].value_counts()
default_counts = loan_data[loan_data['loan_status'] == 1]['loan_amount_bin'].value_counts()

# Calculate the default rate per bin
default_rate = (default_counts / bin_counts) * 100

# Print the default rate per bin
print("Default Rate per Bin:")
print(default_rate)
```
